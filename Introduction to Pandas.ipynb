# SLICING AND FILTERING OF OUR DATAFRAMES

# SLICING

```python
#If we want to filter,slice, and/or index, we can use either the ILOC or LOC
#ILOC - Integer Location: Here you call the areas/parameters making use of an integer or numerical attribute
#LOC - Location: in addition to making use of a numerical attribute, also use a text or string attribute to call parameters
#If i want to print the first record in our dataframe using ILOC
data.iloc[0,:]                                                #This picks the information in the first row 
```

```python
#To call the 3rd record
data.iloc[2,:]
```

```python
#To call the first 2 records ie for Uche and Zane
data.iloc[:2,:]         #This means get all the records for the first 2 rows(index 0,1 excluding 2)
```

```python
#To pick the 3rd and 4th record,ie Patrick and Zane's record
data.iloc[2:,:]       #This means pick everything from row 2 upwards
```

```python
data.iloc[1:3,:]          #iloc also takes range, it sees range as one argument, this will give all the information in rows 1-2 
```

```python
#To pick the 1st, 3rd, and 4th record,ie Ngozi, Patrick and Zane's record
data.iloc[[0,2,3],:]
```

```python
#To pick the 1st, 3rd, and 4th record,ie Ngozi, Patrick and Zane's record
#Multiple items should be wrapped in a list when using iloc 
data.iloc[[0,2,3],:]
```

#  FILTERING

```python
#To get information for people with exam score greater than 32
data['Exam Score'] > 32
```

```python
#To return the above information as a record, copy the above function and insert into the data function
data[data['Exam Score'] > 32]
```

```python
#For multiple filtering
data
```

```python
#To filter out info for clients who have test scores greater than 30 and exam score greater than 32
#For multiple filtering, separate the conditions making use of brackets()
data[(data['Test Score'] > 30) & (data['Exam Score'] > 32)]
```

```python
#In Pandas,AND is denoted as & while OR is denoted as |
# AND = &    Here, both conditions must be true
# OR = |     Here, at least one of the conditions must be met
```

```python
#To filter out info for clients who have test scores greater than 30 or exam score greater than 32

data[(data['Test Score'] > 30) | (data['Exam Score'] > 32)]
```

```python
#Looking at the AnnualIncome column
data.AnnualIncome
```

```python
#The AnnualIncome column is recognized as an object/string due to the prescence of the $ symbol.
#To remove the $ symbol from the items in the AnnualIncome column. we use the 'strip' function
#str.strip is used to remove a particular character from a column
#strip does not have an inplace function, so,to save your answer in memory, assign the column to the strip function argument
data.AnnualIncome = data.AnnualIncome.str.strip('$')
```

```python
#recall the column and we now see that the $ symbol has been removed
data.AnnualIncome
```

```python
data.AnnualIncome.info()
#Even after removing the $ symbol, Python still sees this column as an object/string
```

```python
#We have to remove the commas(,) as well but in this case, the str.strip method did not work
data.AnnualIncome.str.strip(',')
```

```python
#We introduce another function called the'str.replace' method, this method takes in 2 arguments:
#The 1st argument is what you want to replace, the 2nd is what you're replacing it with
data.AnnualIncome.str.replace(",","")          #this tells panda to replace the , in the column with empty value/nothing
```

```python
#To store our values, assign the column to the str.replace function
data.AnnualIncome= data.AnnualIncome.str.replace(",","")
```

```python
#now that both the $ and comma symbols have been removed, we can convert the column using Casting 
#In pandas, casting is done using the astype 
#We will now do casting to convert the column to integer datatype using the 'astype' syntax/function
#We will also assign the column to the casting function to store our results permanently
data.AnnualIncome = data.AnnualIncome.astype(int)
```

```python
data.info()      #Now we see that our AnnualIncome column has become an integer datatype
```

```python
#Treating the Birthdate column to change the data type from object datetime, passing in our desired format as the 2nd argument
#We reassign it to the data.Birthdate column to store our values in memory.
data.BirthDate = pd.to_datetime(data.BirthDate, format = '%d/%m/%Y')
```

```python
data.info()     #The BirthDate column has now been converted to datetime datatype
```
